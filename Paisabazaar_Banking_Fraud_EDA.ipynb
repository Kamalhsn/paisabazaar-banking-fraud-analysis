## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Creating a copy of the dataset to preserve the original data
df_wrangle = df.copy()
````

Based on the inspection of unique values, certain string entries were identified as placeholders rather than valid data.
These values do not carry analytical meaning and can interfere with numerical and statistical operations.
Therefore, they are converted to NaN so that they can be handled consistently as missing values.


````python
# Replacing identified placeholder values with NaN
placeholder_values = ['No Data', 'Not Specified', 'NA', 'N/A', '']

df_wrangle.replace(placeholder_values, np.nan, inplace=True)
````

Pandas and other analytical libraries recognize NaN as missing data.
Converting placeholder text values to NaN ensures that missing information is treated correctly during
statistical analysis, visualization, and further preprocessing steps.


````python
# Removing unwanted characters and converting to numeric where required
numeric_columns = [
    'Annual_Income',
    'Monthly_Inhand_Salary',
    'Outstanding_Debt',
    'Total_EMI_per_month',
    'Amount_invested_monthly',
    'Monthly_Balance'
]

for col in numeric_columns:
    df_wrangle[col] = (
        df_wrangle[col]
        .astype(str)
        .str.replace('_', '', regex=False)
        .str.replace(',', '', regex=False)
    )
    df_wrangle[col] = pd.to_numeric(df_wrangle[col], errors='coerce')
````

````python
# Stripping whitespace from categorical columns
categorical_columns = df_wrangle.select_dtypes(include='object').columns

for col in categorical_columns:
    df_wrangle[col] = df_wrangle[col].str.strip()
````

````python
# Converting negative ages to NaN (invalid values)
df_wrangle.loc[df_wrangle['Age'] < 0, 'Age'] = np.nan
````

#### Chart  1: Distribution of Customer Age


````python
min_age = df_wrangle['Age'].min()
max_age = df_wrangle['Age'].max()
num_customers = df_wrangle.shape[0]

print(f"Minimum Age: {min_age}")
print(f"Maximum Age: {max_age}")
print(f"Total Number of Customers: {num_customers}")
````

````python
# Chart - 1 visualization code

# Calculate IQR for Age
Q1 = df_wrangle['Age'].quantile(0.25)
Q3 = df_wrangle['Age'].quantile(0.75)
IQR = Q3 - Q1

# Calculate bin width using FreedmanDiaconis Rule
# Formula: bin_width = 2 * IQR / (n^(1/3))
n = len(df_wrangle['Age'])
bin_width_fd = 2 * IQR / (n**(1/3))

# Calculate number of bins
age_range = df_wrangle['Age'].max() - df_wrangle['Age'].min()
num_bins_fd = int(np.ceil(age_range / bin_width_fd))

print(f"Calculated optimal number of bins using FreedmanDiaconis Rule: {num_bins_fd}")

plt.figure(figsize=(10, 6))
sns.histplot(df_wrangle['Age'], bins=num_bins_fd, kde=True, color='skyblue')
plt.title('Distribution of Customer Age (FreedmanDiaconis Bins)')
plt.xlabel('Age')
plt.ylabel('Number of Customers')
plt.xticks(np.arange(10, 60, 5)) # Set x-ticks at 5-year intervals for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
````

##### 1. Why did you pick the specific chart?

I chose a histogram for the 'Age' column because it's the most effective chart type for visualizing the distribution of a single numerical variable, like customer age. It helps us understand where most customers' ages are concentrated, how spread out they are, and identify any patterns.

To ensure the best representation, the number of bins for the histogram was determined using the FreedmanDiaconis Rule. This method calculates an optimal bin width that is robust to outliers, providing a more accurate and statistically sound display of the data's underlying shape.

Additionally, a Kernel Density Estimate (KDE) line was overlaid on the histogram. This smooth curve offers a continuous representation of the age distribution, which complements the discrete bars of the histogram by making the overall shape and trends clearer and less affected by the exact binning, thus providing a more refined view of the data's density.

### Chart  5: Credit Utilization Ratio vs Credit Score


What is Credit Utilization Ratio?

It's basically how much of your available credit you're actually using. Think of it as a percentage: (Total Credit Used / Total Available Credit) * 100.

Let's look at some examples:

Example 1: Single Credit Card

You have a credit card with a maximum spending limit of $10,000.

Your current balance (the amount you owe) on that card is $3,000.

Your Credit Utilization Ratio would be: ($3,000 / $10,000) * 100 = 30%.

In this case, you are using 30% of your available credit.

Example 2: Multiple Credit Cards

Card A: Limit = $5,000, Balance = $1,000

Card B: Limit = $10,000, Balance = $4,000

Total Credit Used: $1,000 + $4,000 = $5,000

Total Available Credit: $5,000 + $10,000 = $15,000

Your overall Credit Utilization Ratio would be: ($5,000 / $15,000) * 100 = 33.33%.

Why it matters for your Credit Score (and Chart 5):

Generally, a lower Credit Utilization Ratio is better for your credit score. Many financial experts suggest keeping this ratio below 30%. If you consistently use a high percentage of your available credit, it can signal to lenders that you might be over-reliant on credit or struggling financially, which can lead to a lower credit score.

In Chart 5, you likely saw that customers with 'Good' Credit Scores generally had lower Credit Utilization Ratios, while those with 'Poor' Credit Scores tended to have higher ratios.
